## Test the routing with Google Maps Routing

```python
config = utils.get_config('config.ini')
```

```python
def get_google_length(posa,posb):
    api_key = config['google_maps']['api_key1']
    
    origin = str(posa[0]) + ',' + str(posa[1]) 
    dest = str(posb[0]) + ',' + str(posb[1]) 
    request_string='https://maps.googleapis.com/maps/api/directions/json?origin=' + origin + '&destination=' + dest + '&mode=walking&key=' + api_key
    request = urllib2.Request(request_string)
    response = urllib2.urlopen(request)
    data = json.loads(response.read())
    return float(re.sub("[^0-9.,]","",data['routes'][0]['legs'][0]['distance']['text']))
```

```python
def get_my_length(sg,posa,posb):
    pos0_i,pos1_i = ou.get_closest_nodes(sg,posa,posb)
    path=ou.get_shortest_path(sg,pos0_i,pos1_i)
    return ou.get_path_distance(ou.get_path_roads(sg,path))
```

```python
def generate_random_node_pairs(sg):
    random.seed(6)
    
    nodes = sg.nodes()
    
    goog = []
    nx = []
    for x in range(0,100):
        posa = random.choice(nodes)[::-1]
        posb = random.choice(nodes)[::-1]
        nx_length = get_my_length(sg,posa,posb)
        goog_length = get_google_length(posa,posb)      

        nx.append(nx_length)
        goog.append(goog_length)
    return nx, goog
```

```python
nx,goog = generate_random_node_pairs(sg)
```

```python
fig = plt.figure(figsize=(8,6))

x = np.arange(0,11)
y = np.arange(0,11)
ax1 = fig.add_subplot(111)
ax1.set_xlabel('Google Maps Distance (mi)')
ax1.set_ylabel('PhotoWalk.me Distance (mi)')
ax1.set_title('Shortest Path Distance')
ax1.set_xlim(0,10)
ax1.set_ylim(0,10)
ax1.xaxis.label.set_fontsize(16)
ax1.yaxis.label.set_fontsize(16)
ax1.title.set_fontsize(20)

ax1.scatter(goog,nx,color='red',s=100)
ax1.plot(x,y,lw=1.5)
#plt.savefig('pw_vs_goog.jpg',dpi=300)
```
